# Separate X and Y features
    X Indepandent feature -> YearExperience
    Y Dependent feature -> Salary

```python
salary[['YearsExperience']]
```

```python
X = salary[['YearsExperience']]
Y = salary[['Salary']]
```

```python
X.head()
```

```python
Y.head()
```

# There are no missing value nor any duplicated rows in this dataset

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
pipe = make_pipeline(SimpleImputer(strategy='median')).set_output(transform='pandas')
```

```python
pipe
```

```python
X_pre = pipe.fit_transform(X)
X_pre.head()
```

# Model Building = Linear Regression
    Equation of Line: y = B0 + B1.X
    B0 => y intercept
    B1 => Slope of the line

    ProfitPredicted = B0 + B1*RND

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
# B0 - y intercept
model.intercept_
```

```python
# B1 - slope of the line
model.coef_
```

salarypredicted = 24848.20 + 9449.96*Salary

Salary = 0 salarypredicted = 24848.20

Salary = 1 it will increase by 9449.96 

```python
Y.head()
```

```python
Ypreds = model.predict(X)
Ypreds
```

```python
Ypreds[:5]
```

```python
Y.head()
```

# Evaluation Metrics
    Mean Squared Error
    Mean Absolute Error
    RMSE
    R2 squared- R2 score

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
# Mean squared erro
MSE = mean_squared_error(Y, Ypreds)
# RMSE Root mean squared error
RMSE = MSE**(1/2)
# MAE
MAE = mean_absolute_error(Y, Ypreds)
# r2 score
r2 = r2_score(Y, Ypreds)

print(f"Mean Squared Error: {MSE:.2f}")
print(f"Root Mean Squared Error: {RMSE:.2f}")
print(f"Mean Absolute Error: {MAE:.2f}")
print(f"R2 squared: {r2*100:.2f}%")
```

# R2 squared value is greater than 80%, we can consider this model for final model building

```python
exp_sample = [0,3.4,7.5,11.1,10.5,5,0.5,2.1]
```

```python
preds = []

for x in exp_sample:
    pred = model.predict([[x]]).round(2)
    preds.append(pred)
```
